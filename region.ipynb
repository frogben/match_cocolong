```python
import pandas as pd
```

```python
df=pd.read_csv('data/Customer.csv',usecols=
['CustomerCode','City','Region','IsOrder','IsDeal'])
```

```python
df=df.dropna(subset=['Region'])
df.IsOrder=df.IsOrder.fillna(0)
df.IsDeal=df.IsDeal.fillna(0)
```

```python
df[:3]
```

```python
Region_list=df.Region.unique()
```

```python
df=df.groupby("Region")
region_ratio={}
for region in Region_list:
#     print(i)
    df_region=df.get_group(region)
    all_num=len(df_region)
    deal=df_region.IsDeal.sum()
    ratio=deal/all_num
    print(df_region)
    print('all',all_num)
    print('deal',deal)
    print('ratio',ratio)
    region_ratio[region]=ratio
    print()
    
 
```
